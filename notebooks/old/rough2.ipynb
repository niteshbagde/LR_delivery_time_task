```python
def order_and_pick_time_difference(df:pd.Series):

    df1 = df["Time_Orderd_hour" , "Time_Orderd_minute"]
    df2 = df["Time_Order_picked_hour", "Time_Order_picked_minute"]
    for i in df1:
        
        ohour = i.split(":")[0]
        ominute= i.split(":")[1]

        onormalize_to_mint = int(ohour) * 60 + int(ominute)

        # print(i , onormalize_to_mint)

    for i in df2:
        
        phour = i.split(":")[0]
        pminute= i.split(":")[1]

        pnormalize_to_mint = int(phour) * 60 + int(pminute)

        # print(i , pnormalize_to_mint)
    return pd.Series(pnormalize_to_mint-onormalize_to_mint)
```
