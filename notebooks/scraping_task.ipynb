```python
def get_lat_long(test:df):
    try:
        id = test["ID"]
        rest_latitude = test["Restaurant_latitude"]
        rest_longitude = test["Restaurant_longitude"]

        dest_latitude = test["Delivery_location_latitude"]
        dest_longitude = test["Delivery_location_longitude"]

        return id, rest_latitude, rest_longitude, dest_latitude,dest_longitude

    except:
        print("error in get lat long function")

        
```

```python
def gt_map_url(pt1_lat,pt1_long,pt2_lat,pt2_long):
    maps_url = f"https://www.google.com/maps/dir/{pt1_lat},+{pt1_long}/{pt2_lat},{pt2_long}/@21.3532611,74.6453192,4.67z/data=!4m2!4m1!3e0"
    return maps_url
```

```python
dist_with_ids = dict()

chrome_options = Options()
chrome_options.add_argument('--headless') # add this line to run Chrome in headless mode
driver = webdriver.Chrome(executable_path, chrome_options=chrome_options)

# driver = webdriver.Chrome(executable_path)
# driver.maximize_window()

for i in range(0,len(df)-1):




    def get_dist():
        temp_dist_list=list()
        try:
            for j in range(0,5):
                try:
                    dist = driver.find_element(By.XPATH, f"//*[@id='section-directions-trip-{j}']/div[1]/div[1]/div[1]/div[2]/div").text

                    num_dist = float(dist[0:-3])
                    temp_dist_list.append(num_dist)
                    # print(num_dist)
                except:
                    pass
            return max(temp_dist_list)
        except:
            pass

    driver.get(gt_map_url(get_lat_long(df)[1][i],get_lat_long(df)[2][i],get_lat_long(df)[3][i],get_lat_long(df)[4][i]))

    lat = driver.find_element(By.XPATH,'/html/body/div[3]/div[9]/div[3]/div[1]/div[2]/div/div[3]/div[1]/div[1]/div[2]/div[1]/div/input')
    lat.send_keys(Keys.CONTROL+"a")
    lat.send_keys(Keys.BACKSPACE)
    lat_input = str(get_lat_long(df)[1][i]) + "," + str(get_lat_long(df)[2][i])
    lat.send_keys(lat_input)

    long = driver.find_element(By.XPATH,'/html/body/div[3]/div[9]/div[3]/div[1]/div[2]/div/div[3]/div[1]/div[2]/div[2]/div[1]/div/input')
    long.send_keys(Keys.CONTROL+"a")
    long.send_keys(Keys.BACKSPACE)
    long_input = str(get_lat_long(df)[3][i]) + "," + str(get_lat_long(df)[4][i])
    long.send_keys(long_input)
    long.send_keys(Keys.RETURN)
    time.sleep(4)

    dist_with_ids[get_lat_long(df)[0][i]] = get_dist()

    # update values in the 'Profession' column
    df.loc[i, 'onground_dist'] = get_dist()


    print("iteration=", i , "dist=",get_dist(), "URL: ", gt_map_url(get_lat_long(df)[1][i],get_lat_long(df)[2][i],get_lat_long(df)[3][i],get_lat_long(df)[4][i]))
    print("completed")

driver.close()

with open("distWithIDs.txt", "w") as f:
    for i in dist_with_ids:
        one = i +":"+ str(dist_with_ids[i])+","+"\n"
        f.writelines(one)
    print("complete")  
    
```
